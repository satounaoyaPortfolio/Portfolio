```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
#csvファイルを読み込む
df = pd.read_csv("data.csv")
#xとyの題名をつける
sns.scatterplot(data=df,x="homerun",y="strikeout")
#日本語の表記を扱えるようにする
sns.set(font="Hiragino Maru Gothic pro")
#セリーグとパリーグの色分けの作業
for f in df['league'].unique():
    plt.scatter(df.loc[df.league == f ,'homerun'], df.loc[df.league == f ,'strikeout'], label=f)
plt.title('三振とホームランの関係')
plt.legend()
plt.show()
```
